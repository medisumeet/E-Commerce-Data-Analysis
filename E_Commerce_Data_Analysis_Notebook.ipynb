```python
df['Revenue'] = 0  # initialize with 0

df.loc[df['IsReturned'] == 0, 'Revenue'] = df['Quantity'] * df['PricePerUnit']
```

```python
revenue_per_category = df[df['Revenue'] != 0].groupby('ProductCategory')['Revenue'].sum()
category_contribution = (revenue_per_category / revenue_per_category.sum()) * 100

plt.pie(category_contribution, labels=category_contribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Product Category Contribution to Revenue')
plt.axis('equal')
plt.show()
```

```python
revenue_per_location=df[df['Revenue'] != 0].groupby('CustomerLocation')['Revenue'].sum()
Location_contribution=(revenue_per_location/revenue_per_category.sum())*100
plt.pie(Location_contribution,labels=Location_contribution.index,autopct='%1.1f%%',startangle=140)
plt.title('location contribution to revenue')
plt.axis('equal')
plt.show()
```
